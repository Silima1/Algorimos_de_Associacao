# Analise de censo trabalhadores com base OneR/algoritmo por regras

```python
import Orange
```

```python
base_censo = Orange.classification.Table('../Dados/Bases_de_dados/census_regras.csv')
```

```python
base_censo.domain
```

```python
base_dividida = Orange.evaluation.testing.sample(base_censo, n=0.25)
base_dividida
```

```python
base_dividida[0]
```

```python
base_dividida[1]
```

```python
base_treinamento = base_dividida[1]
```

```python
base_teste = base_dividida[0]
```

```python
len(base_treinamento), len(base_teste)
```

```python
treino = Orange.classification.rules.CN2Learner()
treinar_regras = treino(base_treinamento)
treinar_regras
```

```python
for regras in treinar_regras.rule_list:
    print(regras)
```

```python
previsao = Orange.evaluation.testing.TestOnTestData(base_treinamento, base_teste, [lambda testdata: treinar_regras])
previsao
```

```python
Orange.evaluation.CA(previsao)
```
