# Classificacao de credito com base Majority learner

```python
# De referir que este algoritmo n√£o tem objetivo de treinar o modelo, mas definir uma base minima pontuacao 
import Orange
base_credit = Orange.data.Table('../Dados/Bases_de_dados/credit_data_regras.csv')
```

```python
base_credit.domain
```

```python
Majority = Orange.classification.MajorityLearner() # variavel para classificacao
Previsao = Orange.evaluation.testing.TestOnTestData(base_credit, base_credit, [Majority])
Orange.evaluation.CA(Previsao)
```

```python
for registo in base_credit: # para visualizacao de classes 0 e 1
    print(registo.get_class())
```
